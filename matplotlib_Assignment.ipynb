```python
import matplotlib.pyplot as plt
```

Part1: Line Plot

```python
x = [1, 2, 3, 4, 5]
y = [10, 15, 25, 30, 50]
plt.figure()
plt.title("Simple Line Chart")
plt.plot(x,y,label="Trend",marker="o")
plt.xlabel("x-values")
plt.ylabel("y-values")
plt.legend()
plt.grid()
plt.show()
```

Part2:Bar chart

```python
students = ['John', 'Jane', 'Alice', 'Bob']
marks = [75, 85, 60, 90]
colors=["red","blue","black","yellow"]
plt.figure()
plt.bar(students,marks,color=colors)
plt.title("Marks obtained by Students")
plt.xlabel("Students")
plt.ylabel("Marks")
plt.show()
```

Part3: Pie Chart

```python
regions = ['North America', 'Europe', 'Asia', 'Others']
revenue = [45, 25, 20, 10]
explode=[]
#finding the maximum salary index
maximum=max(revenue)
val=0
for i in range(len(revenue)):
  if revenue[i]==maximum:
    print(i)
    val=i
print(val)
explode=[]
for i in range(len(revenue)):
  if i==val:
    explode.append(0.08)
  else:
    explode.append(0)
print(explode)
plt.figure()
plt.pie(revenue,labels=regions,autopct="%1.1f%%",explode=explode)
plt.title("Revenue Distribution")
plt.show()
```

Part4: Histogram

```python
import numpy as np
x=np.random.randint(1,100,1000)
plt.figure()
plt.title("Frequency of Numbers from  1-100")
plt.xlabel("No. of groups")
plt.ylabel("Frequency")
plt.hist(x,bins=10,color="blue",edgecolor="black")
plt.show()
```
