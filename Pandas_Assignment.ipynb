```python
import pandas as pd
```

```python
#DataFrame Creation
data = {
      'Name': ['Alice', 'Bob', 'Charlie', 'Diana', 'Eve'],
      'Age': [24, 27, 22, 32, 29],
      'Department': ['HR', 'Finance', 'IT', 'Marketing', 'HR'],
      'Salary': [45000, 54000, 50000, 62000, 47000]
  }
#PRINTING FIRST FIVE DATAPOINTS
df=pd.DataFrame(data)
print(df.head())
```

```python
# Get the summary statistics of the 'Age' and 'Salary' columns.

X=df.select_dtypes(include=["number"])
print(X)
#Printing Stats
print(X.info())
print(X.describe())

```

```python
#Average Salary of Employees in HR Department
Average_salary=df[df["Department"]=="HR"]["Salary"].mean()
print(Average_salary)
```

```python
#Adding New Colimn "Bonus"
df["Bonus"]=df["Salary"]*0.10
print(df)
```

```python
#Filter the DataFrame to show employees aged between 25 and 30.
df_filtered=df[(df["Age"]<30) & (df["Age"]>25)]
print(df_filtered)
```

```python
#Grouping data
group_data=df.groupby("Department")
print(group_data["Salary"].mean())
```

```python
#Sorting data
sorted_df=df.sort_values("Salary")
print(sorted_df)
sorted_df.to_csv("Sorted_data.csv")
```
