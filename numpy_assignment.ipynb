```python
import numpy as np
```

Part1: Numpy 1D array operations


```python
list1=[]
for i in range(1,21):
  list1.append(i)
print(list1)
array1=np.array(list1)
print(array1)
```

```python
#Performing Mathematical Operations
#Sum
summation=np.sum(array1)
print(summation)
mean_val=np.mean(array1)
print(mean_val)
median_val=np.median(array1)
print(median_val)
std_val=np.std(array1)
print(std_val)
```

```python
#Getting indices of the array
for i in range(len(array1)):
  if array1[i]>10:
    print(i)
```

Part2:2D-Array

```python
arr=np.arange(1,17)
arr=arr.reshape(4,4)
print(arr)

```

```python
#Transpose of array
print(np.transpose(arr))
```

Part3: 3X3 Array

```python
#Creating 1st Array
x1=np.random.randint(1,21,size=(3,3))
print(x1)
#Part a
#Creating 2nd Array
x2=np.random.randint(1,21,size=(3,3))
print(x2)

#Element wise Addition
Add_final=np.add(x1,x2)
print(f"The Addition of two matrices:\n",Add_final)

#Element wise Subtraction
Sub_final=np.subtract(x1,x2)
print(f"The Subtraction of two matrices:\n",Sub_final)

#Element wise Multiplication
mul_final=np.multiply(x1,x2)
print(f"The Multiplication of two matrices:\n",mul_final)

#Part b:
dot_final=np.dot(x1,x2)
print(f"The dot product of two matrices:\n",dot_final)
```

Part4:Reshaping Array


```python
#Creating 1D array
arr_new=np.arange(1,13)
print(arr_new)
print(arr_new.shape)

#Reshaping Array
arr_reshaped=arr_new.reshape(3,4)
print(arr_reshaped)

#Slicing First two rows
print(arr_reshaped[0:2,:])

#Slicing last two columns
print(arr_reshaped[:,2:4])

print(arr_reshaped[arr_reshaped>6])
```
